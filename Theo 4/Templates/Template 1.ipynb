# Random Numbers in Python
In statistical physics, we work a lot with random numbers. This notebook contains some basic examples for generating random numbers and for plotting probability distributions in Python.

In most numerical examples, we will import the 'numpy' module for mathematical functions and the 'matplotlib' module for plotting results. 

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Sampling random numbers
The 'random' module of numpy has many build-in functions for generating random numbers, which are distributed according to common probability distributions. Here are some basic examples for sampling and displaying random numbers:

### Uniform and normal distribution

As two basic examples, we generate $N_{sample}=10$ random numbers, which are (i) uniformly distributed in the intervall $[0,1]$ and (ii) distributed according to the normal distribution $\mathcal{N}(0,1)$, which means a Gaussian probability distribution with zero mean and a variance of 1.

Those distributions can be generated by the numpy functions 'numpy.random.rand' and 'numpy.random.randn', respectively. 

```python
samples=10 # number of random samples

# uniform distribution for interval [0,1]:
numbers_uniform=np.random.rand(samples)

print(numbers_uniform)
```

```python
# normal distribution N(0,1):
numbers_normal=np.random.randn(samples)

print(numbers_normal)
```

Note that for these small sample sizes, the mean value and standard deviation of this set of numnbers will still deviate significantly from the exact values:

```python
print(numbers_normal.mean())
print(numbers_normal.std())
```

#### Tips: 

To generate similar distributions over different intervals or with different mean values and variance, the numbers generated in this way can be rescaled by multiplying them by a common factor and shifted by adding a common offset value. 

Check out the documentation for the numpy.random module for other built-in distribution funtions. 


### Histrogram plots:
To visualize the distributions we can use histogram plots. Specifically, we can use the argument 'density=True' to display histograms as properly normalized probability distributions.

```python
samples=1000 # number of random samples

# normal distribution N(0,1):
numbers_normal=np.random.randn(samples)

# excat distribution:
x_vec=np.linspace(-3,3,201)
dist_normal=1/np.sqrt(2*np.pi)*np.exp(-x_vec**2/2)

n_bins = 40
plt.figure(dpi = 100)
plt.plot(x_vec,dist_normal)
plt.hist(numbers_normal, bins=n_bins,density=True)
plt.xlabel(r'$x$ ', fontsize = 20)
plt.ylabel(r'$P(x)$ ', fontsize =20)
plt.tick_params(axis='both', which='major', labelsize=16)
plt.show()
```
