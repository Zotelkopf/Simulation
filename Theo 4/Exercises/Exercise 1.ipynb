### 0. Imports and Generators

```python
import numpy as np
import matplotlib.pyplot as plt
rng = np.random.default_rng()
dist_normal = lambda x, mu, sigma: 1 / (np.sqrt(2 * np.pi) * sigma) * \
                                   np.exp(-(x - mu)**2 / (2 * sigma**2))
```

### 1. Gau√üverteilung

```python
sample_size_normal = 1000
mean_normal = 5
std_normal = 2
```

```python
sample_normal = rng.normal(mean_normal, std_normal, sample_size_normal)
sample_normal[:5]
```

```python
bin_count = 50
_, bins, _ = plt.hist(sample_normal, bins=bin_count, density=True)
x = np.linspace(bins[0], bins[-1], 1000)
plt.plot(x, dist_normal(x, mean_normal, std_normal), color='red', lw=2)
plt.xlabel(r'$x$', fontsize=14)
plt.ylabel(r'$p(x)$', fontsize=14)
plt.show()
```

### 2. Zentraler Grenzwertsatz

```python
N = 1000
M = 100
sample_size_uniform = (N, M)
sample_uniform = rng.uniform(1, 3, sample_size_uniform)
sample_uniform[:5, 0]
```

```python
mean_uniform = sample_uniform.mean()
std_uniform = sample_uniform.std(ddof=1)
mean_uniform, std_uniform
```

```python
sample_sum = sample_uniform.sum(axis=1)
sample_sum[:5]
```

```python
bin_count = 50
_, bins, _ = plt.hist(sample_sum, bins=bin_count, density=True)
x = np.linspace(bins[0], bins[-1], 1000)
plt.plot(x, dist_normal(x, M*mean_uniform, np.sqrt(M)*std_uniform), color='red', lw=2)
plt.xlabel(r'$x$', fontsize=14)
plt.ylabel(r'$p(x)$', fontsize=14)
plt.show()
```

### 3. Binomialverteilung

```python
N = 10000
p = 0.4
sample_size_binomial = 1000
```

```python
sample_binomial = rng.binomial(N, p, sample_size_binomial)
sample_binomial[:5]
```

```python
bin_count = 50
_, bins, _ = plt.hist(sample_binomial, bins=bin_count, density=True)
x = np.linspace(bins[0], bins[-1], 1000)
plt.plot(x, dist_normal(x, p*N, np.sqrt(N*p*(1-p))), color='red', lw=2)
plt.xlabel(r'$N_1$', fontsize=14)
plt.ylabel(r'$p(N_1)$', fontsize=14)
plt.show()
```
